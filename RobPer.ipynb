# RobPer Periodograms

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Paso 1: Leer los datos del archivo periodicidades.txt
file_path = 'resultados_periodogramas_100periodos_modelsine.txt'  # Cambia esto por la ruta correcta de tu archivo
periods_days = pd.read_csv(file_path, header=None, names=['period_days'])

# Paso 2: Convertir las periodicidades de días a minutos
periods_minutes = periods_days['period_days']# * 24 * 60  # 1 día = 1440 minutos
periods_days['period_minutes'] = periods_minutes


# Paso 3: Cálculo de estadísticas básicas
mean_minutes = periods_minutes.mean()
median_minutes = periods_minutes.median()
moda_minutes = periods_minutes.mode()
std_minutes = periods_minutes.std()
min_minutes = periods_minutes.min()
max_minutes = periods_minutes.max()

print(f"Media de periodicidades: {mean_minutes:.2f} minutos")
print(f"Mediana de periodicidades: {median_minutes:.2f} minutos")

#if len(moda_minutes) > 1:
modas_formateadas = ', '.join([f"{moda:.2f}" for moda in moda_minutes])
print(f"Modas de periodicidades: {modas_formateadas} minutos")
#else:
#    print(f"Moda de periodicidades: {moda_minutes.values[0]:.2f} minutos")

print(f"Desviación estándar: {std_minutes:.2f} minutos")
print(f"Periodicidad mínima: {min_minutes:.2f} minutos")
print(f"Periodicidad máxima: {max_minutes:.2f} minutos")

# Paso 4: Generación de gráficos

# Crear una figura con varios subplots
plt.figure(figsize=(16, 12))

# Gráfico 1: Histograma
plt.subplot(2, 2, 1)
sns.histplot(periods_minutes, bins=40, kde=True, color='blue')
plt.axvline(mean_minutes, color='red', linestyle='--', label=f'Media: {mean_minutes:.2f} min')
plt.axvline(median_minutes, color='orange', linestyle='--', label=f'Mediana: {median_minutes:.2f} min')
plt.title('Histograma de Periodicidades (en minutos)')
plt.xlabel('Periodicidad (minutos)')
plt.ylabel('Frecuencia')
#plt.xlim(0, 400)  # Limitar el eje x
plt.legend()

# Gráfico 2: KDE (Kernel Density Estimate)
plt.subplot(2, 2, 2)
sns.kdeplot(periods_minutes, fill=True, color='green')
plt.axvline(mean_minutes, color='red', linestyle='--', label=f'Media: {mean_minutes:.2f} min')
plt.axvline(median_minutes, color='orange', linestyle='--', label=f'Mediana: {median_minutes:.2f} min')
plt.title('Gráfico de Densidad KDE de Periodicidades (en minutos)')
plt.xlabel('Periodicidad (minutos)')
plt.ylabel('Densidad')
#plt.xlim(-1000, 1000)  # Limitar el eje x
plt.legend()

# Gráfico 3: Boxplot
plt.subplot(2, 2, 3)
sns.violinplot(x=periods_minutes, color='orange')
plt.axvline(mean_minutes, color='red', linestyle='--', label=f'Media: {mean_minutes:.2f} min')
plt.axvline(median_minutes, color='orange', linestyle='--', label=f'Mediana: {median_minutes:.2f} min')
plt.title('Boxplot de Periodicidades (en minutos)')
plt.xlabel('Periodicidad (minutos)')
#plt.xlim(0, 200)  # Limitar el eje x
plt.legend()



# Gráfico 4: Gráfico de dispersión (scatter plot) con tendencia
# Calcular los cuartiles
Q1 = periods_minutes.quantile(0.25)
Q3 = periods_minutes.quantile(0.75)
IQR = Q3 - Q1

# Definir los límites inferior y superior
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar los outliers
periods_minutes_sin_outliers = periods_minutes[(periods_minutes >= limite_inferior) & (periods_minutes <= limite_superior)]

# Graficar la tendencia sin outliers
plt.subplot(2, 2, 4)
sns.regplot(x=np.arange(len(periods_minutes_sin_outliers)), y=periods_minutes_sin_outliers, scatter=True, line_kws={"color": "red"})
plt.title('Tendencia de las Periodicidades (en minutos)')
plt.xlabel('Índice de muestra')
plt.ylabel('Periodicidad (minutos)')
plt.ylim(0, 200)  # Limitar el eje y a 200 minutos
plt.legend()

# Ajustar los gráficos
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo, ignorando los valores NA
file_path = 'resultados_periodogramas_1_300periodos_modelsine.txt'
data = pd.read_csv(file_path, sep='\t', header=None)  # Ajusta el separador si es necesario
periods_minutes = data[0].dropna()  # Asume que la periodicidad está en la primera columna

# Convertir a numpy array para trabajar con los datos
periods_minutes = periods_minutes.to_numpy()

# Función bootstrap para calcular estadísticos
def bootstrap(data, statistic_func, num_samples=1000):
    n = len(data)
    stats = np.empty(num_samples)
    for i in range(num_samples):
        sample = np.random.choice(data, size=n, replace=True)
        stats[i] = statistic_func(sample)
    return stats

# Calcular la media, mediana, moda, desviación estándar y sus intervalos de confianza
def calcular_estadisticas(data, num_bootstrap=1000):
    # Media
    bootstrap_means = bootstrap(data, np.mean, num_samples=num_bootstrap)
    mean = np.mean(data)
    mean_ci = np.percentile(bootstrap_means, [2.5, 97.5])
    
    # Mediana
    bootstrap_medians = bootstrap(data, np.median, num_samples=num_bootstrap)
    median = np.median(data)
    median_ci = np.percentile(bootstrap_medians, [2.5, 97.5])
    
    # Moda
    mode = pd.Series(data).mode().values[0]  # Calcular moda como serie
    bootstrap_modes = bootstrap(data, lambda x: pd.Series(x).mode().values[0], num_samples=num_bootstrap)
    mode_ci = np.percentile(bootstrap_modes, [2.5, 97.5])

    # Desviación estándar
    bootstrap_stds = bootstrap(data, np.std, num_samples=num_bootstrap)
    std = np.std(data)
    std_ci = np.percentile(bootstrap_stds, [2.5, 97.5])
    
    return {
        "mean": (mean, mean_ci),
        "median": (median, median_ci),
        "mode": (mode, mode_ci),
        "std": (std, std_ci)
    }

# Calcular estadísticos con bootstrap
estadisticas = calcular_estadisticas(periods_minutes)

# Mostrar resultados
print(f"Media de periodicidades: {estadisticas['mean'][0]:.2f} minutos")
print(f"Intervalo de confianza de la media: [{estadisticas['mean'][1][0]:.2f}, {estadisticas['mean'][1][1]:.2f}] minutos\n")

print(f"Mediana de periodicidades: {estadisticas['median'][0]:.2f} minutos")
print(f"Intervalo de confianza de la mediana: [{estadisticas['median'][1][0]:.2f}, {estadisticas['median'][1][1]:.2f}] minutos\n")

print(f"Moda de periodicidades: {estadisticas['mode'][0]:.2f} minutos")
print(f"Intervalo de confianza de la moda: [{estadisticas['mode'][1][0]:.2f}, {estadisticas['mode'][1][1]:.2f}] minutos\n")

print(f"Desviación estándar: {estadisticas['std'][0]:.2f} minutos")
print(f"Intervalo de confianza de la desviación estándar: [{estadisticas['std'][1][0]:.2f}, {estadisticas['std'][1][1]:.2f}] minutos\n")

# Graficar los resultados con los intervalos de confianza
plt.figure(figsize=(10, 6))

# Graficar media con IC
plt.errorbar(0, estadisticas['mean'][0], yerr=[[estadisticas['mean'][0] - estadisticas['mean'][1][0]], [estadisticas['mean'][1][1] - estadisticas['mean'][0]]],
             fmt='o', label=f'Media ({estadisticas["mean"][0]:.2f} min)', capsize=5)

# Graficar mediana con IC
plt.errorbar(1, estadisticas['median'][0], yerr=[[estadisticas['median'][0] - estadisticas['median'][1][0]], [estadisticas['median'][1][1] - estadisticas['median'][0]]],
             fmt='o', label=f'Mediana ({estadisticas["median"][0]:.2f} min)', capsize=5)

# Graficar moda con IC
plt.errorbar(2, estadisticas['mode'][0], yerr=[[estadisticas['mode'][0] - estadisticas['mode'][1][0]], [estadisticas['mode'][1][1] - estadisticas['mode'][0]]],
             fmt='o', label=f'Moda ({estadisticas["mode"][0]:.2f} min)', capsize=5)

# Graficar desviación estándar con IC
plt.errorbar(3, estadisticas['std'][0], yerr=[[estadisticas['std'][0] - estadisticas['std'][1][0]], [estadisticas['std'][1][1] - estadisticas['std'][0]]],
             fmt='o', label=f'Desviación Estándar ({estadisticas["std"][0]:.2f} min)', capsize=5)

plt.xticks([0, 1, 2, 3], ['Media', 'Mediana', 'Moda', 'Desviación Estándar'])
plt.ylabel('Valor (minutos)')
plt.title('Estadísticos de Periodicidades con Intervalos de Confianza')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Paso 1: Leer los datos del archivo periodicidades.txt
file_path = 'resultados_periodogramas_1_300periodos_modelsine.txt'  # Cambia esto por la ruta correcta de tu archivo
periods_days = pd.read_csv(file_path, header=None, names=['period_days'])

# Paso 2: Convertir las periodicidades de días a minutos
periods_minutes = periods_days['period_days']  # * 24 * 60  # 1 día = 1440 minutos
periods_minutes = periods_minutes.dropna()  # Filtrar valores NaN
periods_days['period_minutes'] = periods_minutes

# Paso 3: Cálculo de estadísticas básicas
mean_minutes = periods_minutes.mean()
median_minutes = periods_minutes.median()
moda_minutes = periods_minutes.mode()
std_minutes = periods_minutes.std()
min_minutes = periods_minutes.min()
max_minutes = periods_minutes.max()

# Intervalo de confianza de la media usando la t-distribución (95%)
conf_interval_mean = stats.t.interval(0.95, df=len(periods_minutes)-1, loc=mean_minutes, scale=std_minutes / np.sqrt(len(periods_minutes)))

# Intervalo de confianza de la mediana usando bootstrap
n_bootstraps = 10000
bootstrapped_medians = np.median(np.random.choice(periods_minutes, (n_bootstraps, len(periods_minutes))), axis=1)
conf_interval_median = np.percentile(bootstrapped_medians, [2.5, 97.5])

# Imprimir los resultados
print(f"Media de periodicidades: {mean_minutes:.2f} minutos")
print(f"Mediana de periodicidades: {median_minutes:.2f} minutos")
modas_formateadas = ', '.join([f"{moda:.2f}" for moda in moda_minutes])
print(f"Modas de periodicidades: {modas_formateadas} minutos")
print(f"Desviación estándar: {std_minutes:.2f} minutos")
print(f"Periodicidad mínima: {min_minutes:.2f} minutos")
print(f"Periodicidad máxima: {max_minutes:.2f} minutos")
print(f"Intervalo de confianza de la media: {conf_interval_mean}")
print(f"Intervalo de confianza de la mediana: {conf_interval_median}")

# Paso 4: Generación de gráficos

# Crear una figura con varios subplots
plt.figure(figsize=(16, 12))

# Gráfico 1: Histograma
plt.subplot(2, 2, 1)
sns.histplot(periods_minutes, bins='auto', kde=True, color='blue')
plt.axvline(mean_minutes, color='red', linestyle='--', label=f'Media: {mean_minutes:.2f} min')
plt.axvline(median_minutes, color='orange', linestyle='--', label=f'Mediana: {median_minutes:.2f} min')
plt.fill_betweenx([0, plt.ylim()[1]], conf_interval_mean[0], conf_interval_mean[1], color='red', alpha=0.2, label='IC de la media (95%)')
plt.fill_betweenx([0, plt.ylim()[1]], conf_interval_median[0], conf_interval_median[1], color='orange', alpha=0.2, label='IC de la mediana (95%)')
plt.title('Histograma de Periodicidades (en minutos)')
plt.xlabel('Periodicidad (minutos)')
plt.ylabel('Frecuencia')
plt.legend()

# Gráfico 2: KDE (Kernel Density Estimate)
plt.subplot(2, 2, 2)
sns.kdeplot(periods_minutes, fill=True, color='green')
plt.axvline(mean_minutes, color='red', linestyle='--', label=f'Media: {mean_minutes:.2f} min')
plt.axvline(median_minutes, color='orange', linestyle='--', label=f'Mediana: {median_minutes:.2f} min')
plt.fill_betweenx([0, plt.ylim()[1]], conf_interval_mean[0], conf_interval_mean[1], color='red', alpha=0.2, label='IC de la media (95%)')
plt.fill_betweenx([0, plt.ylim()[1]], conf_interval_median[0], conf_interval_median[1], color='orange', alpha=0.2, label='IC de la mediana (95%)')
plt.title('Gráfico de Densidad KDE de Periodicidades (en minutos)')
plt.xlabel('Periodicidad (minutos)')
plt.ylabel('Densidad')
plt.legend()

# Gráfico 3: Boxplot
plt.subplot(2, 2, 3)
sns.violinplot(x=periods_minutes, color='orange')
plt.axvline(mean_minutes, color='red', linestyle='--', label=f'Media: {mean_minutes:.2f} min')
plt.axvline(median_minutes, color='orange', linestyle='--', label=f'Mediana: {median_minutes:.2f} min')
plt.title('Boxplot de Periodicidades (en minutos)')
plt.xlabel('Periodicidad (minutos)')
plt.legend()

# Gráfico 4: Gráfico de dispersión (scatter plot) con tendencia
# Calcular los cuartiles
Q1 = periods_minutes.quantile(0.25)
Q3 = periods_minutes.quantile(0.75)
IQR = Q3 - Q1

# Definir los límites inferior y superior
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar los outliers
periods_minutes_sin_outliers = periods_minutes[(periods_minutes >= limite_inferior) & (periods_minutes <= limite_superior)]

# Graficar la tendencia sin outliers
plt.subplot(2, 2, 4)
sns.regplot(x=np.arange(len(periods_minutes_sin_outliers)), y=periods_minutes_sin_outliers, scatter=True, line_kws={"color": "red"})
plt.title('Tendencia de las Periodicidades (en minutos)')
plt.xlabel('Índice de muestra')
plt.ylabel('Periodicidad (minutos)')
plt.ylim(0, 200)  # Limitar el eje y a 200 minutos
plt.legend()

# Agregar líneas horizontales para cada moda
for moda in moda_minutes:
    plt.axhline(y=moda, color='blue', linestyle='--', label=f'Moda: {moda:.2f} min')

# Evitar etiquetas duplicadas en la leyenda
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

# Ajustar los gráficos
plt.tight_layout()
plt.show()
```
